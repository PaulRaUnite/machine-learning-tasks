Алгоритм рішення

```python
import numpy as np
from scipy.spatial.distance import cdist

from itertools import count
def connect(points: pd.DataFrame) -> pd.DataFrame:
    a = points.to_numpy()
    return pd.DataFrame(cdist(a, a, 'euclidean'))

def clusterize(m: pd.DataFrame, thresold: float) -> np.array:
    clusters = np.zeros(m.shape[0])
    free = set(range(m.shape[0]))

    cluster_num = count(1)
    while len(free) > 0:
        seed = free.pop()
        n = next(cluster_num)

        queue = [seed]
        while len(queue) > 0:
            p = queue.pop(0)
            assert clusters[p] == 0 or clusters[p] == n
            clusters[p] = n

            for x, v in enumerate(m.loc[p]):
                if x!= p and x in free and 0 <= v <= thresold:
                    free.remove(x)
                    queue.append(x)
    return clusters

def remove_more_than(m: pd.DataFrame, thresold: float) -> pd.DataFrame:
     m = m.copy()
     for i in range(m.shape[0]):
        for j in range(m.shape[1]):
            if m.loc[i,j] >= thresold:
                m.loc[i,j] = -1
     return m
```

```python
d = connect(df)
Rs = np.arange(0, 3, step=0.1)
C = []
for r in Rs:
    C.append(len(set(clusterize(d, r))))
```
